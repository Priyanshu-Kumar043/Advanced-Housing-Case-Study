```python
## Data Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

## display all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
pd.pandas.set_option('display.max_rows',None)
```

## Missing Values

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
for features in features_with_na:
    print(features, np.round(dataset[features].isnull().mean(),4)*100," %missing values")
```

### Find relation between missing value and SalePrice

```python
for features in features_with_na:
    data = dataset.copy()
    
    data[features] = np.where(data[features].isnull(),1,0)
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
```

```python
print("No of Id : {}".format(len(dataset.Id)))
```

```python
numerical_features = [features for features in dataset.columns if dataset[features].dtypes != 'O']
print("No of Numerical Features : {}".format(len(numerical_features)))

dataset[numerical_features].head()
```

```python
## List of variables containing year info
year_feature = [feature for feature in dataset[numerical_features].columns if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for features in year_feature:
    print(features,dataset[features].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price vs Year Sold")
```

```python
year_feature
```

```python
## here we will compare each year feauture with SalePrice

for features in year_feature:
        if features != 'YrSold':
            data = dataset.copy()
            data[features] = data['YrSold'] - data[features]
            sns.scatterplot(data=data,x=data[features],y=data['SalePrice'],color='y')
            plt.xlabel(features)
            plt.ylabel('SalePrice')
            plt.show()
```

```python
## Discrete feature and continuous features
discrete_features = [features for features in numerical_features if len(dataset[features].unique())<25 and features not in year_feature]
print("Number of Discrete features:",len(discrete_features))
```

```python
discrete_features
```

```python
dataset[discrete_features].head()
```

```python
## Relation between Numerical Features and SalePrice

for features in discrete_features:
    data = dataset.copy()
    data.groupby(features)['SalePrice'].median().plot.bar(color = 'y')
    plt.xlabel(features)
    plt.ylabel('SalePrice')
    plt.title(features)
    plt.show()
    
```

```python
## Continuous features

continuous_features = [features for features in numerical_features if features not in discrete_features and features not in year_feature and features != 'Id']
print("Number of Continuous Feature:",len(continuous_features))
```

```python
continuous_features
```

```python
## Analysing continuous feature using histogram

for features in continuous_features:
    data = dataset.copy()
    sns.histplot(data[features],bins=30,kde=True,)
    plt.xlabel(features)
    plt.ylabel('Count')
    plt.title(features)
    plt.show()
```

## Exploratory Data Analysis Part 2

```python
## We will be using logarithmic transformation to remove skew

for features in continuous_features:
        data = dataset.copy()
        if 0 in data[features].unique():
            pass
        else:
            data[features] = np.log(data[features])
            data['SalePrice'] = np.log(data['SalePrice'])
            plt.scatter(data[features],data['SalePrice'],color='g')
            plt.xlabel(features)
            plt.ylabel("SalePrice")
            plt.title(features)
            plt.show()
```

## Outliers

```python
for features in continuous_features:
        data = dataset.copy()
        if 0 in data[features].unique():
            pass
        else:
            data[features] = np.log(data[features])
            data['SalePrice'] = np.log(data['SalePrice'])
            data.boxplot(column=features)
            plt.xlabel(features)
            plt.ylabel("SalePrice")
            plt.title(features)
            plt.show()
            
```

```python
categorical_features = [features for features in dataset.columns if features not in numerical_features]
print("Number of Categorical Features:",len(categorical_features))
```
