```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# For feature selection
from sklearn.linear_model import Lasso,Ridge
from sklearn.feature_selection import SelectFromModel

pd.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv("X_train Advance House Price Prediction.csv")
```

```python
dataset.head()
```

```python
# Capture the depent feature
y_train = dataset[['SalePrice']]
y_train
```

```python
X_train = dataset.drop(['Id','SalePrice'],axis=1)
```

```python
x_train = X_train.copy()
```

```python
X_train
```

```python
### Apply Feature Selection
# First, specify Ridge Regression Model
# Select a suitable alpha (equivalent of penalty)
# The bigger the alpha the less feature will be selected

# Then use SelectFromModel object from sklearn, which
# will select the features having non-zero coefficients

feature_sel_model = SelectFromModel(Ridge(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_feat_Ridge = X_train.columns[(feature_sel_model.get_support())]

print('Total Features: {}'.format((X_train.shape[1])))
print('Selected Features: {}'.format(len(selected_feat_Ridge)))
```

```python
selected_feat_Ridge
```

```python
### Apply Feature Selection
# First, specify Lasso Regression Model
# Select a suitable alpha (equivalent of penalty)
# The bigger the alpha the less feature will be selected

# Then use SelectFromModel object from sklearn, which
# will select the features having non-zero coefficients

feature_sel_model = SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_feat_Lasso = X_train.columns[(feature_sel_model.get_support())]

print('Total Features: {}'.format((X_train.shape[1])))
print('Selected Features: {}'.format(len(selected_feat_Lasso)))

```

```python
selected_feat_Lasso
```
